```python
import dl_translate as dlt

mt = dlt.TranslationModel("nllb200")
```

```python
text = "Meta AI has built a single AI model capable of translating across 200 different languages with state-of-the-art quality."

# The new translation is much faster than before
%time print(mt.translate(text, source=dlt.lang.nllb200.ENGLISH, target=dlt.lang.nllb200.FRENCH))
```

```python
# Sanskrit is now available (previously unavailable)
print(mt.translate(text, source=dlt.lang.nllb200.ENGLISH, target=dlt.lang.nllb200.SANSKRIT))
```

```python
# Sicilian is now available (previously unavailable)
print(mt.translate(text, source=dlt.lang.nllb200.ENGLISH, target=dlt.lang.nllb200.SICILIAN))
```

```python
# Yue Chinese is now available (previously unavailable)
print(mt.translate(text, source=dlt.lang.nllb200.ENGLISH, target=dlt.lang.nllb200.YUE_CHINESE))
```
